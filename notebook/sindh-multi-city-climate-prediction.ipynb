```python

# Project B: Sindh Multi-City Climate Prediction System

# Importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout


# # Loading Original Data

# df = pd.read_csv("/content/1958 to 2025.csv", skiprows=3)
# # Skipping 3 rows because first 3 rows contain metadata
# # (geographical/location information not needed for time-series forecasting)

# print(df.shape)
# print(df.columns)
# print(df.info())

# #Selecting Required Features

# # CMCC_CM2_VHR4 selected as representative high-resolution climate model
# # to maintain consistency and avoid inter-model variability

# df = df[[
#     "date",
#     "temperature_2m_mean_CMCC_CM2_VHR4 (°C)",
#     "wind_speed_10m_mean_CMCC_CM2_VHR4 (km/h)",
#     "relative_humidity_2m_mean_CMCC_CM2_VHR4 (%)",
#     "precipitation_sum_CMCC_CM2_VHR4 (mm)"
# ]]

# # Rename columns to easier names
# df = df.rename(columns={
#     "temperature_2m_mean_CMCC_CM2_VHR4 (°C)": "temp",
#     "wind_speed_10m_mean_CMCC_CM2_VHR4 (km/h)": "wind",
#     "relative_humidity_2m_mean_CMCC_CM2_VHR4 (%)": "humidity",
#     "precipitation_sum_CMCC_CM2_VHR4 (mm)": "rain"
# })

# print(df.head())
# print(df.columns)

# #Cleaning Data

# # Fix column names
# df.columns = df.columns.str.strip()

# # Convert date safely
# df["date"] = pd.to_datetime(df["date"], errors="coerce")

# # Remove invalid date rows
# df = df[df["date"].notna()]

# # Sort by date
# df = df.sort_values("date").reset_index(drop=True)

# # Filter years 2000-2024
# df = df[df["date"].dt.year.between(2000, 2024)]

# print(df.shape)
# print(df.head())
# print(df.info())

# #Adding City Column = "Karachi"

# df["city"] = "Karachi"
# print(df.head())

# #Checking Missing Values

# print("Missing values per column:")
# print(df.isnull().sum())

# #Checking Duplicates

# print("Number of duplicate rows:", df.duplicated().sum())

# #Outlier Analysis
# #Not removing outliers because extreme weather is real and valid

# num_cols = df.select_dtypes(include=["number"]).columns

# report = []

# for col in num_cols:

#     Q1 = df[col].quantile(0.25)
#     Q3 = df[col].quantile(0.75)
#     IQR = Q3 - Q1

#     lower = Q1 - 1.5 * IQR
#     upper = Q3 + 1.5 * IQR

#     outliers = df[(df[col] < lower) | (df[col] > upper)]

#     report.append({
#         "Column": col,
#         "Min": df[col].min(),
#         "Max": df[col].max(),
#         "Mean": df[col].mean(),
#         "Lower_Bound": lower,
#         "Upper_Bound": upper,
#         "Outlier_Count": len(outliers),
#         "Outlier_%": round((len(outliers) / len(df)) * 100, 2)
#     })

# outlier_report_df = pd.DataFrame(report)
# outlier_report_df = outlier_report_df.sort_values(by="Outlier_Count", ascending=False)

# print("\nOUTLIER SUMMARY REPORT")
# print(outlier_report_df)

# # Karachi Histogram

# num_cols = df.select_dtypes(include=["number"]).columns

# for col in num_cols:
#     plt.figure(figsize=(6, 4))
#     plt.hist(df[col].dropna(), bins=30, edgecolor="black")
#     plt.title(f"Distribution of {col}")
#     plt.xlabel(col)
#     plt.ylabel("Frequency")
#     plt.show()

# # Saving Karachi Data

# df.to_csv("cleaned_climate_data.csv", index=False)
# print("Saved: cleaned_climate_data.csv")


# # Loading Karachi Data


# df = pd.read_csv("/content/cleaned_climate_data.csv")

# # Defining 15 cities of Karachi

# cities = [
#     "Karachi", "Hyderabad", "Sukkur", "Larkana", "Jacobabad",
#     "Dadu", "Thatta", "Badin", "Mirpurkhas", "Tharparkar",
#     "Shikarpur", "Ghotki", "Umerkot", "Sanghar", "Khairpur"
# ]

# # Temperature bias per city based on real climate knowledge
# city_temp_bias = {
#     "Jacobabad": 6,
#     "Tharparkar": 5.5,
#     "Sukkur": 5,
#     "Shikarpur": 4.5,
#     "Larkana": 4,
#     "Khairpur": 3,
#     "Ghotki": 2.8,
#     "Dadu": 2.5,
#     "Hyderabad": 1.5,
#     "Mirpurkhas": 1.2,
#     "Umerkot": 0.8,
#     "Sanghar": 0.7,
#     "Karachi": 0,
#     "Thatta": -0.5,
#     "Badin": -1,
# }

# # Noise strength per city
# city_noise_strength = {
#     "Jacobabad": 1.8,
#     "Tharparkar": 1.7,
#     "Sukkur": 1.6,
#     "Shikarpur": 1.5,
#     "Larkana": 1.4,
#     "Khairpur": 1.2,
#     "Ghotki": 1.1,
#     "Dadu": 1.0,
#     "Hyderabad": 0.9,
#     "Mirpurkhas": 0.8,
#     "Umerkot": 0.7,
#     "Sanghar": 0.6,
#     "Karachi": 0.5,
#     "Thatta": 0.4,
#     "Badin": 0.3,
# }

# # Generative Multi-City Data


# all_data = []
# np.random.seed(42)

# for city in cities:

#     temp_df = df.copy()
#     temp_df["city"] = city

#     bias = city_temp_bias.get(city, 0)
#     noise = city_noise_strength.get(city, 0.5)

#     # Temperature
#     temp_df["temp"] += bias + np.random.normal(0, noise, len(temp_df))

#     # Humidity
#     temp_df["humidity"] = 70 - (temp_df["temp"] * 0.8) \
#                           + np.random.normal(0, noise * 2, len(temp_df))
#     temp_df["humidity"] = np.clip(temp_df["humidity"], 0, 100)

#     # Rain
#     temp_df["rain"] = temp_df["humidity"] * 0.02 \
#                       + np.random.normal(0, noise * 0.3, len(temp_df))
#     temp_df["rain"] = np.clip(temp_df["rain"], 0, None)

#     # Wind
#     if "wind" in temp_df.columns:
#         temp_df["wind"] = temp_df["wind"] + np.random.normal(0, noise * 0.3, len(temp_df))
#         temp_df["wind"] = np.clip(temp_df["wind"], 0, None)

#     all_data.append(temp_df)

# # Combine all cities
# df_multi = pd.concat(all_data, ignore_index=True)

# print("FINAL SHAPE:", df_multi.shape)
# print(df_multi["city"].value_counts())
# print(df_multi.sample(10))

# # MISSING VALUE ANALYSIS (-999)
# # Inject -999 values to simulate missing/invalid measurements
# np.random.seed(42)
# df_multi_missing = df_multi.copy()

# for city in cities:
#     city_mask = df_multi_missing["city"] == city
#     city_indices = df_multi_missing[city_mask].index
#     # inject -999 in 5% of rows per city
#     missing_indices = np.random.choice(
#         city_indices,
#         size=int(len(city_indices) * 0.05),
#         replace=False
#     )
#     df_multi_missing.loc[missing_indices, "temp"] = -999

# # Count -999 per city
# missing_counts = {}
# for city in cities:
#     city_data = df_multi_missing[df_multi_missing["city"] == city]["temp"]
#     missing_counts[city] = (city_data == -999).sum()

# missing_series = pd.Series(missing_counts)

# # Bar chart: distribution of -999 values per city
# plt.figure(figsize=(12, 5))
# missing_series.sort_values(ascending=False).plot(
#     kind="bar",
#     color="tomato",
#     edgecolor="black"
# )
# plt.title("Distribution of -999 (Missing) Values Per City")
# plt.xlabel("City")
# plt.ylabel("Count of Missing Values")
# plt.xticks(rotation=45)
# plt.tight_layout()
# plt.show()

# # Imputation: replace -999 with per-city mean temperature
# for city in cities:
#     city_mask = df_multi_missing["city"] == city
#     city_mean = df_multi_missing.loc[
#         city_mask & (df_multi_missing["temp"] != -999), "temp"
#     ].mean()
#     df_multi_missing.loc[
#         city_mask & (df_multi_missing["temp"] == -999), "temp"
#     ] = city_mean

# # Verify no -999 values remain
# print("Missing -999 values remaining after imputation:")
# print((df_multi_missing["temp"] == -999).sum())

# # Use imputed data going forward
# df_multi = df_multi_missing.copy()

# # CORRELATION ANALYSIS

# numeric_df = df_multi.select_dtypes(include=["float64", "int64"])
# corr_matrix = numeric_df.corr()

# print("Correlation with Temperature:")
# print(corr_matrix["temp"].sort_values(ascending=False))

# # Heatmap
# plt.figure(figsize=(10, 7))
# sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
# plt.title("Feature Correlation Matrix (Climate Dataset)")
# plt.show()

# # Bar plot
# plt.figure(figsize=(8, 5))
# corr_matrix["temp"].drop("temp").sort_values().plot(kind="bar")
# plt.title("Feature Correlation with Temperature")
# plt.ylabel("Correlation")
# plt.grid(alpha=0.3)
# plt.show()

# # STATISTICAL SUMMARY

# print(df_multi.describe())

# # Histograms for all features
# num_cols = df_multi.select_dtypes(include=["number"]).columns

# for col in num_cols:
#     plt.figure(figsize=(8, 4))
#     plt.hist(df_multi[col].dropna(), bins=30, edgecolor="black")
#     plt.title(f"Distribution of {col}")
#     plt.xlabel(col)
#     plt.ylabel("Frequency")
#     plt.show()

# # City-wise mean comparison
# city_stats = df_multi.groupby("city")[num_cols].mean()
# print(city_stats)


# # SAVE MULTI-CITY DATA

# df_multi.to_csv("df_multi.csv", index=False)
# print("Saved: df_multi.csv")

# # BLOCK 18: CITY-WISE HYPERPARAMETER TABLE
# # Different cities get different hyperparameters based on climate complexity
# # Hotter/more variable cities get more LSTM units and longer lookback

city_hyperparams = {
    "Karachi":    {"LSTM_Units": 64,  "Lookback": 7,  "Epochs": 15},
    "Hyderabad":  {"LSTM_Units": 64,  "Lookback": 7,  "Epochs": 15},
    "Sukkur":     {"LSTM_Units": 128, "Lookback": 10, "Epochs": 20},
    "Larkana":    {"LSTM_Units": 128, "Lookback": 10, "Epochs": 20},
    "Jacobabad":  {"LSTM_Units": 128, "Lookback": 14, "Epochs": 25},
    "Dadu":       {"LSTM_Units": 64,  "Lookback": 7,  "Epochs": 15},
    "Thatta":     {"LSTM_Units": 64,  "Lookback": 7,  "Epochs": 15},
    "Badin":      {"LSTM_Units": 64,  "Lookback": 7,  "Epochs": 15},
    "Mirpurkhas": {"LSTM_Units": 64,  "Lookback": 10, "Epochs": 15},
    "Tharparkar": {"LSTM_Units": 128, "Lookback": 14, "Epochs": 25},
    "Shikarpur":  {"LSTM_Units": 128, "Lookback": 10, "Epochs": 20},
    "Ghotki":     {"LSTM_Units": 64,  "Lookback": 10, "Epochs": 15},
    "Umerkot":    {"LSTM_Units": 64,  "Lookback": 7,  "Epochs": 15},
    "Sanghar":    {"LSTM_Units": 64,  "Lookback": 7,  "Epochs": 15},
    "Khairpur":   {"LSTM_Units": 128, "Lookback": 10, "Epochs": 20},
}

hyperparams_df = pd.DataFrame(city_hyperparams).T
hyperparams_df.index.name = "City"

print("\nCity-wise Hyperparameter Comparison Table:")
print(hyperparams_df)

# Loading Data for Training

LOOKBACK = 10
TARGET = "temp"

df = pd.read_csv("/content/df_multi.csv")

# SEQUENCE CREATION FUNCTION

def create_sequences(data, lookback):
    X, y = [], []
    for i in range(len(data) - lookback):
        X.append(data[i:i+lookback])
        y.append(data[i+lookback])
    return np.array(X), np.array(y)

# MODEL BUILDING FUNCTION


def build_model(lstm_units, lookback):
    model = Sequential([
        LSTM(lstm_units, return_sequences=True, input_shape=(lookback, 1)),
        Dropout(0.2),
        LSTM(lstm_units // 2),
        Dropout(0.2),
        Dense(1)
    ])
    model.compile(optimizer="adam", loss="mse")
    return model

# TRAIN MODEL PER CITY

results = []
city_predictions = {}

for city in df["city"].unique():

    print(f"\nTraining model for: {city}")

    city_df = df[df["city"] == city].copy()
    city_df = city_df.sort_values("date")

    # USE CITY HYPERPARAMETERS
    params = city_hyperparams.get(city, {
        "LSTM_Units": 64,
        "Lookback": 10,
        "Epochs": 15
    })

    LSTM_UNITS = params["LSTM_Units"]
    LOOKBACK = params["Lookback"]
    EPOCHS = params["Epochs"]

    values = city_df[[TARGET]].values

    scaler = MinMaxScaler()
    scaled = scaler.fit_transform(values)

    # LOOKBACK IS NOW DYNAMIC
    X, y = create_sequences(scaled, LOOKBACK)

    split = int(len(X) * 0.8)
    X_train, X_test = X[:split], X[split:]
    y_train, y_test = y[:split], y[split:]

    X_train = X_train.reshape(X_train.shape[0], LOOKBACK, 1)
    X_test = X_test.reshape(X_test.shape[0], LOOKBACK, 1)

    # MODEL NOW USES CITY LSTM SIZE
    model = build_model(LSTM_UNITS, LOOKBACK)

    # CITY-SPECIFIC TRAINING EPOCHS
    model.fit(
        X_train, y_train,
        epochs=EPOCHS,
        batch_size=32,
        verbose=1
    )

    pred = model.predict(X_test)

    y_test_inv = scaler.inverse_transform(y_test)
    pred_inv = scaler.inverse_transform(pred)

    rmse = np.sqrt(mean_squared_error(y_test_inv, pred_inv))
    r2 = r2_score(y_test_inv, pred_inv)

    results.append([city, rmse, r2])

    city_predictions[city] = {
        "actual": y_test_inv.flatten(),
        "predicted": pred_inv.flatten()
    }

# RESULTS TABLE

results_df = pd.DataFrame(results, columns=["City", "RMSE", "R2"])

results_df = results_df.sort_values(
    by=["RMSE", "R2"],
    ascending=[True, False]
)

print("\nModel Performance - All 15 Cities:")
print(results_df.to_string(index=False))

best_city = results_df.iloc[0]["City"]
worst_city = results_df.iloc[-1]["City"]

print("\nBest Predictable City:", best_city)
print("Worst Predictable City:", worst_city)

#BLOCK 24: CROSS-CITY ANALYSIS

print("\n" + "="*50)
print("CROSS-CITY ANALYSIS")
print("="*50)

print(f"\nEasiest City to Predict: {best_city}")
print(f"Reason: {best_city} has stable and consistent climate patterns")
print(f"with low temperature variance, making it easier for LSTM to learn.")

print(f"\nHardest City to Predict: {worst_city}")
print(f"Reason: {worst_city} experiences extreme temperatures and")
print(f"high seasonal variability, making predictions less accurate.")

print("\nTop 5 Easiest Cities to Predict:")
print(results_df.head(5)[["City", "RMSE", "R2"]].to_string(index=False))

print("\nTop 5 Hardest Cities to Predict:")
print(results_df.tail(5)[["City", "RMSE", "R2"]].to_string(index=False))

# VALIDATION GRAPHS (3 CITIES)

cities_to_plot = ["Karachi", "Hyderabad", "Jacobabad"]

plt.figure(figsize=(18, 5))

for i, city in enumerate(cities_to_plot):

    actual = city_predictions[city]["actual"]
    predicted = city_predictions[city]["predicted"]

    plt.subplot(1, 3, i + 1)
    plt.plot(actual, label="Actual")
    plt.plot(predicted, label="Predicted")
    plt.title(f"{city} - Actual vs Predicted")
    plt.xlabel("Time")
    plt.ylabel("Temperature (°C)")
    plt.legend()

plt.tight_layout()
plt.show()


# ACCURACY REPORT (R2 AND RMSE ALL 15 CITIES)

print("\nACCURACY REPORT - R2 and RMSE for All 15 Cities:")
print(results_df[["City", "RMSE", "R2"]].to_string(index=False))

# OVERALL RMSE HEATMAP

plt.figure(figsize=(10, 6))
sns.heatmap(
    results_df.set_index("City")[["RMSE"]],
    annot=True,
    fmt=".3f",
    cmap="coolwarm",
    linewidths=0.5
)
plt.title("City-wise Prediction Error (RMSE)")
plt.tight_layout()
plt.show()

# SEASONAL ERROR HEATMAP

def get_season(month):
    if month in [3, 4]:
        return "Spring"
    elif month in [5, 6, 7]:
        return "Summer"
    elif month in [8, 9]:
        return "Monsoon"
    else:
        return "Winter"

seasonal_rmse = {}

for city in df["city"].unique():

    city_df = df[df["city"] == city].copy()
    city_df = city_df.sort_values("date")
    city_df["date"] = pd.to_datetime(city_df["date"])
    city_df["month"] = city_df["date"].dt.month
    city_df["season"] = city_df["month"].apply(get_season)

    actual = city_predictions[city]["actual"]
    predicted = city_predictions[city]["predicted"]

    # get test portion dates
    test_dates = city_df.iloc[int(len(city_df) * 0.8):].reset_index(drop=True)
    test_dates = test_dates.iloc[:len(actual)].reset_index(drop=True)

    season_rmse_city = {}

    for season in ["Spring", "Summer", "Monsoon", "Winter"]:
        idx = test_dates[test_dates["season"] == season].index
        if len(idx) > 0:
            rmse_s = np.sqrt(mean_squared_error(
                actual[idx],
                predicted[idx]
            ))
            season_rmse_city[season] = round(rmse_s, 3)
        else:
            season_rmse_city[season] = np.nan

    seasonal_rmse[city] = season_rmse_city

seasonal_df = pd.DataFrame(seasonal_rmse).T
seasonal_df.index.name = "City"

print("\nSeasonal RMSE per City:")
print(seasonal_df)

plt.figure(figsize=(12, 8))
sns.heatmap(
    seasonal_df,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5
)
plt.title("Prediction Error (RMSE) Across Cities and Seasons")
plt.xlabel("Season")
plt.ylabel("City")
plt.tight_layout()
plt.show()

```
